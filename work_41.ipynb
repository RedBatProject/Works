```python
performance = pd.DataFrame()
smp = []
mdl = []
nump = 0
numd = len(sampler.keys())*len(models.keys())
print(numd)
for sm in sampler.keys():
  data = data2.sample(frac=1,random_state=2342)
  rus = sampler[sm]()


  # X_,Y_= rus.fit_resample(data, data['Machine failure'])
  # print(Y_.sum()/y_.shape[0])

  # X_['Machine failure'] = Y_
  # X_ = data
  # X_ = X_.sample(frac=1)
  train = data.sample(frac=0.8,random_state=2342)
  test = data.drop(train.index)
  train,tr= rus.fit_resample(train, train['Machine failure'])
  train['Machine failure'] = tr

  # test = pd.concat([test[test['Machine failure']==1].sample(frac=1),test[test['Machine failure']==0].sample(test[test['Machine failure']==0].shape[0])])
  X,Y,x,y = train.drop('Machine failure',axis=1).to_numpy(),train['Machine failure'].to_numpy(),test.drop('Machine failure',axis=1).to_numpy(),test['Machine failure'].to_numpy()
  for mod in models.keys():
    nump +=1
    print(nump/numd,sm,mod)
    # try:
    #   model = models[mod](random_state=1234)
    # except:
    model = models[mod]()
    model.fit(X,Y)
    y_ = model.predict(x)
    smp.append(sm)
    mdl.append(mod)
    performance[f'{sm} {mod}'] = [accuracy_score(y,y_),precision_score(y,y_,average='macro'),recall_score(y,y_,average='macro'),f1_score(y,y_,average='macro')]

performance.index = ['accuracy','precision','recall','f1']
performanceT = performance.T
performanceT['model'] = mdl
performanceT['sampler'] = smp
# performanceT.sort_values(by='f1',ascending=False)
```

```python
def fitness_function(solution):
    max_depth = solution[0]
    n_estimator = solution[1]
    learning_rate = solution[2]
    clf = models['XGB'](max_depth=int(max_depth),
                    n_estimators=int(n_estimator),random_state=0,learning_rate=learning_rate,
                    )
    # Fit the model
    clf.fit(X, Y)
    # Make the predictions
    y_ = clf.predict(x)
    # cm = confusion_matrix(y,y_)
    # km = cm[0][1]+cm[1][0]
    return f1_score(y,y_,average='macro')

problem = {
    "fit_func": fitness_function,
    "lb": [8, 100, 0.01],
    "ub": [12, 1000, .9],
    "minmax": "max",
}
```
